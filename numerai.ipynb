```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import  mean_absolute_error
import numpy as np
from xgboost import XGBClassifier

training_data = pd.read_csv("https://numerai-public-datasets.s3-us-west-2.amazonaws.com/latest_numerai_training_data.csv.xz")
uncorr1 = training_data.columns[training_data.columns.str.startswith('feature_int')]
uncorr2 = training_data.columns[training_data.columns.str.startswith('feature_cha')]
uncorr3 = training_data.columns[training_data.columns.str.startswith('feature_str')]
uncorr4 = training_data.columns[training_data.columns.str.startswith('feature_dex')]
uncorr5 = training_data.columns[training_data.columns.str.startswith('feature_con')]
uncorr6 = training_data.columns[training_data.columns.str.startswith('feature_wis')]
#feature_cols = training_data.columns[training_data.columns.str.startswith('feature')]
```

```
import random
def randomfeatures(lists):
  sampling1 = random.sample(lists, k=3)
  sampling2 = random.sample(lists, k=4) 
  sampling3 = random.sample(lists, k=5)
  sampling4 = random.sample(lists, k=6)
  total = [sampling1, sampling2, sampling3, sampling4]
  return random.choice(total)

```

# Backtesting different models

```
from scipy.stats import spearmanr
from xgboost import XGBClassifier
storage = {}

for i in range(1000):
  X1 = training_data[uncorr1].sample(axis='columns')
  X2 = training_data[uncorr2].sample(axis='columns')
  X3 = training_data[uncorr3].sample(axis='columns')
  X4 = training_data[uncorr4].sample(axis='columns')
  X5 = training_data[uncorr5].sample(axis='columns')
  X6 = training_data[uncorr6].sample(axis='columns')
  trains = randomfeatures([X1, X2, X3, X4, X5, X6])
  
  X = pd.concat(trains, axis=1)
  Y = training_data['target']

  trainX, testX, trainY, testY = train_test_split(X, Y, random_state=50)
  trainY = [str(z) for z in trainY]
  model = XGBClassifier(objective ='reg:squarederror', random_state=50)
  fitted = model.fit(trainX, trainY)
  result = model.predict(testX)


  result = [float(z) for z in result]
  coef, p = spearmanr(testY, result)

  columns = X.columns.values
  storage[str(columns)] = coef
  dicts = {k: v for k, v in sorted(storage.items(), key=lambda item: item[1], reverse=True)}
  value_at_index = list(dicts.values())[0], list(dicts.keys())[0]
  print(value_at_index, i)

```

```
from sklearn.ensemble import RandomForestClassifier
storage = {}

for i in range(100):

  X1 = training_data[uncorr1].sample(axis='columns')
  X2 = training_data[uncorr2].sample(axis='columns')
  X3 = training_data[uncorr3].sample(axis='columns')
  X4 = training_data[uncorr4].sample(axis='columns')
  X5 = training_data[uncorr5].sample(axis='columns')
  X6 = training_data[uncorr6].sample(axis='columns')

  trains = randomfeatures([X1, X2, X3, X4, X5, X6])
  
  X = pd.concat(trains, axis=1)
  Y = training_data['target']

  trainX, testX, trainY, testY = train_test_split(X, Y, random_state=1)
  trainY = [str(z) for z in trainY]
  model = RandomForestClassifier(random_state=1, n_estimators= 75)
  fitted = model.fit(trainX, trainY)
  result = model.predict(testX)


  result = [float(z) for z in result]
  coef, p = spearmanr(testY, result)

  columns = X.columns.values
  storage[str(columns)] = coef
  dicts = {k: v for k, v in sorted(storage.items(), key=lambda item: item[1], reverse=True)}
  value_at_index = list(dicts.values())[0], list(dicts.keys())[0]
  print(value_at_index, i)
```

# Submissions

```
feature = ['feature_charisma46', 'feature_constitution15', 'feature_dexterity5', 'feature_strength34']
```

```
from sklearn.ensemble import RandomForestClassifier

X = training_data[feature]
Y = training_data['target']
trainX, testX, trainY, testY = train_test_split(X, Y, random_state=1)
trainY = [str(z) for z in trainY]
model = RandomForestClassifier(random_state=1, n_estimators= 150, max_depth=4, max_leaf_nodes=15, min_samples_split=5)

fitted = model.fit(trainX, trainY)
result = model.predict(testX)
result = [float(z) for z in result]
print(len(result), result.count(0.5))
error = mean_absolute_error(testY, result)
print(error)
```

```
from xgboost import XGBClassifier

X = training_data[feature]
Y = training_data['target']
trainX, testX, trainY, testY = train_test_split(X, Y, random_state=10)
trainY = [str(z) for z in trainY]
model = XGBClassifier(objective ='reg:squarederror', random_state=5)

fitted = model.fit(trainX, trainY)
result = model.predict(testX)
result = [float(z) for z in result]
print(len(result), result.count(0.5))
error = mean_absolute_error(testY, result)
print(error)
```
